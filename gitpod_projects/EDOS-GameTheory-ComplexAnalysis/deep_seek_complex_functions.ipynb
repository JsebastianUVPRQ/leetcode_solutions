```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D

# Define the complex function
def f(z):
    return z**2

```

```python
# Generate a grid of complex numbers
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Apply the function to the grid
W = f(Z)

# 3D Surface Plot
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(131, projection='3d')
ax.plot_surface(X, Y, np.abs(W), cmap='viridis', edgecolor='none')
ax.set_title('3D Surface Plot of |f(z)|')
ax.set_xlabel('Re(z)')
ax.set_ylabel('Im(z)')
ax.set_zlabel('|f(z)|')

# Riemann Sphere Representation
def riemann_sphere(z):
    x = 2 * np.real(z) / (1 + np.abs(z)**2)
    y = 2 * np.imag(z) / (1 + np.abs(z)**2)
    z = (-1 + np.abs(z)**2) / (1 + np.abs(z)**2)
    return x, y, z

X_r, Y_r, Z_r = riemann_sphere(W)

ax = fig.add_subplot(132, projection='3d')
ax.plot_surface(X_r, Y_r, Z_r, facecolors=plt.cm.hsv(np.angle(W) / (2 * np.pi)), rstride=1, cstride=1, alpha=0.8)
ax.set_title('Riemann Sphere Representation')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Field of Vectors Colored by Magnitude
ax = fig.add_subplot(133)
magnitude = np.abs(W)
angle = np.angle(W)
ax.quiver(X, Y, np.real(W/magnitude), np.imag(W/magnitude), magnitude, cmap='viridis', angles='xy', scale_units='xy', scale=1, width=0.005)
ax.set_title('Field of Vectors Colored by Magnitude')
ax.set_xlabel('Re(z)')
ax.set_ylabel('Im(z)')

plt.tight_layout()
plt.show()
```
